```python
import pandas as pd
import requests
from io import StringIO
```

```python
orig_url = 'https://drive.google.com/file/d/1Gq9beJ456xaTC_v6iwVSEQ-R5Q0UViUH/view?usp=sharing'

file_id = orig_url.split('/')[-2]
dwn_url = 'https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
df_stollen = pd.read_csv(csv_raw, skiprows=6, sep='";"', names=['Desc', '2010', '2011', '2012', '2013', '2014', '2015', 
                                                                '2016', '2017', '2018', '2019'])
df_stollen
```

```python
df_stollen = df_stollen.replace({'"':''}, regex=True)
df_stollen['Desc'] = df_stollen['Desc'].replace({'Diefstal van fiets':'Stollen bikes'}, regex=True)
df_stollen
```

```python
total_stollen = df_stollen.loc[df_stollen['Desc'] == 'Stollen bikes']
total_stollen = total_stollen[['Desc', '2016', '2017', '2018', '2019']]
total_stollen
```
