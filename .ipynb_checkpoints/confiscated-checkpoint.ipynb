```python
import pandas as pd
import requests
from io import StringIO
```

```python
orig_url = 'https://drive.google.com/file/d/1jxoeEgw3fq-qD6ph1SKQSHD7fYyM1f4S/view?usp=sharing'

file_id = orig_url.split('/')[-2]
dwn_url = 'https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
df_confiscated = pd.read_csv(csv_raw, skiprows=2, sep='\t', nrows=9, names=['Desc', '2010', '2016', '2017', '2018', '2019'])
df_confiscated
```

```python
df_confiscated['Desc'] = df_confiscated['Desc'].replace({'totaal':'Confiscated bikes'}, regex=True)
conf = df_confiscated.loc[df_confiscated['Desc'] == 'Confiscated bikes']
conf = conf.drop('2010', 1)
conf[['2016', '2017', '2018', '2019']] = conf[['2016', '2017', '2018', '2019']].astype('int64')
conf
```

```python
df_left = conf.copy()
df_left['Desc'] = df_left['Desc'].replace({'Confiscated bikes':'Conf.bikes left market'}, regex=True)
df_left[['2016', '2017', '2018', '2019']] = round(df_left[['2016', '2017', '2018', '2019']].apply(lambda x: x/3)).astype('int64')
df_left
```

```python
total_conf = pd.concat([conf, df_left])
total_conf
```
