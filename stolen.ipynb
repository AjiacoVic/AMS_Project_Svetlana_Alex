```python
import pandas as pd
import requests
from io import StringIO
```

```python
orig_url = 'https://drive.google.com/file/d/1Gq9beJ456xaTC_v6iwVSEQ-R5Q0UViUH/view?usp=sharing'

file_id = orig_url.split('/')[-2]
dwn_url = 'https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
df_stolen = pd.read_csv(csv_raw, skiprows=6, sep='";"', names=['Desc', '2010', '2011', '2012', '2013', '2014', '2015', 
                                                                '2016', '2017', '2018', '2019'])
df_stolen
```

```python
df_stolen = df_stolen.replace({'"':''}, regex=True)
df_stolen['Desc'] = df_stolen['Desc'].replace({'Diefstal van fiets':'Stolen bikes'}, regex=True)
df_stolen
```

```python
total_stolen = df_stolen.loc[df_stolen['Desc'] == 'Stolen bikes']
total_stolen = total_stolen[['Desc', '2016', '2017', '2018', '2019']]
total_stolen
```
